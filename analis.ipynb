```python
from pandas.io.json import json_normalize
import pandas as pd
import matplotlib.pyplot as plt
import geopandas
```

```python
file = 'tuvamedee.txt'
f = open(file, encoding='utf8')
user = [eval(us) for us in f]
temp_df = json_normalize(user)
```

```python
df = pd.DataFrame(temp_df, columns = ['id','first_name','last_name','deactivated', 'is_closed', 'can_access_closed','sex', 'online',
                                   'bdate', 'city.title', 'country.title', 'can_post', 'can_see_all_posts',
                                  'can_see_audio','can_write_private_message','site', 'last_seen.time',
                                  'last_seen.platform', 'university_name', 'faculty_name', 'schools.name',
                                  'skype','status', 'facebook','twitter','instagram', 'photo_200_orig'])
```

```python
df.count()
```

```python
df.to_excel("tuvamedee.xlsx", encoding='utf8') #save to excel
```

```python
df['deactivated'] = df['deactivated'].fillna('active')
df.groupby('deactivated')['id'].count().plot.pie(autopct='%1.0f%%',label='Статус пользователей')
df = df.loc[df['deactivated'] == 'active']
```
